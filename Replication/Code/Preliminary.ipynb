```python
# Set Up
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data
dt_castle = pd.read_stata('../Data/castle-doctrine-2000-2010.dta')
dt_homicide = pd.read_stata('../Data/homicide-1960-2010.dta')
```

```python
# dt_castle.columns
# np.mean(dt_castle['robbery'])
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.hist(dt_castle[dt_castle['year'] == 2005]['homicide'], 
         bins=30, alpha=0.6, label='2005')

plt.hist(dt_castle[dt_castle['year'] == 2010]['homicide'], 
         bins=30, alpha=0.6, label='2010')

plt.xlabel('Homicide (per 100,000 population)')
plt.ylabel('Frequency')
plt.title('Distribution of Homicide in 2005 vs 2010')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
